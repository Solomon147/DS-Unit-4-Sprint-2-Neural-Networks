## 1. Define the following terms:

- **Neuron:**A Neuron can be defined as a function that receives at least one weighted input and sums them to make an activated output. Additionally they receive inputs and pass on their signal to the next layer of the neural network. The Neuron consits of the an input layer, hidden layer and an output layer. The Neuron bascially consist of numbers between 0 to 1.

- **Input Layer:** The input layer is the first layer of the neural network. It will receive at least one input, and sometimes many. Number of inputs nodes is basically correlated with the difficulty or complexity of the task at hand. After receiving the inputs, this layer will update the weights and send information to the hidden layer. This layer could also be referred to as the visible layer.

- **Hidden Layer:** They're used in neural network and can be picked based on experience. We really do not see their values and that's why they're referred to as hidden(invisible). They're also referred to as the blackbox. They're either connected to another hidden layer or the output layer

- **Output Layer:** The job of the output layer is to produce an output. This is what is been predicted. There would be 1 node when it is a binary classification or a simple regression. While in a multi class regression problem, there will be many nodes as possible outputs

- **Activation:** This can be referred to as a transfer function. Each node in a neural network has one and they're usually the same within each layer of the network. It determines whether a node sends information on to connected nodes in deeper layers.


- **Backpropagation:** This is the means by which neural network improves, it tells the ANN whether its prediction was correct or incorrect, and the network adjusts its parameters accordingly. Typically, it is done by computing the negative gradient of the error and then going through the network backwards to adjust the weights

### Multilayer Perceptron <a id="Q3"></a>

Using the sample candy dataset, implement a Neural Network Multilayer Perceptron class that uses backpropagation to update the network's weights. Your Multilayer Perceptron should be implemented in Numpy. 
Your network must have one hidden layer.

Once you've trained your model, report your accuracy. Explain why your MLP's performance is considerably better than your simple perceptron's on the candy dataset. 

```python
class NeuralNetwork:
    def __init__(self, inputs, hiddenNodes, outputNodes=1):
        self.inputs = inputs
        self.hiddenNodes =  hiddenNodes
        self.outputNodes = outputNodes
        self.inputWeights = np.random.randn(self.inputs, self.hiddenNodes)
        self.hiddenWeights = np.random.randn(self.hiddenNodes, self.outputNodes)
    def sigmoid(self, s):
        return 1 / (1+np.exp(-s))
    def sigmoidPrime(self, s):
        return s * (1-s)
    def feed_forward(self, X):
        self.hiddenSum = np.dot(X, self.inputWeights)
        self.hiddenActv = self.sigmoid(self.hiddenSum)
        self.outputSum = np.dot(self.hiddenActv, self.hiddenWeights)
        self.outputActv = self.sigmoid(self.outputSum)
        return(self.outputActv) # used as output for backward
    def backward(self, X, y, output, learning_rate=.1):
        self.outputError = y - output
        self.outputDelta = learning_rate * (self.outputError * self.sigmoidPrime(output)) 
        self.hiddenError = self.outputDelta.dot(self.hiddenWeights.T)
        self.hiddenDelta = self.hiddenError * self.sigmoidPrime(self.hiddenActv)
        self.hiddenWeights += self.hiddenActv.T.dot(self.outputDelta)
        self.inputWeights += X.T.dot(self.hiddenDelta)
    def train(self, X, y, epochs=100, learning_rate=.1):
        for _ in range(epochs):
            output = self.feed_forward(X)
            self.backward(X, y, output, learning_rate=learning_rate)
        self.loss = np.mean(np.square(y - self.feed_forward(X)))
        print(f'\nLoss after {epochs} epochs was {self.loss}\n')
    def predict(self, X):
        output = self.feed_forward(X)
        predictions = []
        for i in output:
            if i[0] >= 0.5:
                predictions.append([1])
            else:
                predictions.append([0])
        self.predictions = np.array(predictions)
    def check(self, y):
        correct_predictions = np.sum(self.predictions == y)
        total_predictions = len(self.predictions)
        accuracy = correct_predictions / total_predictions
        return(f'The model had a {accuracy} accuracy score')
```

```python
#creates an instance of neural network class
nn = NeuralNetwork(inputs=2, hiddenNodes=8)
# trains network and gets loss
nn.train(X, y, epochs=10000, learning_rate=.01)
# creates predictions
nn.predict(X)
# checks the models accuracy
nn.check(y)
```

This multilayer perception works better because it can adapt to the non-linearity of the data in a way that the perceptron could not. Could not get a perfect score because some kids were brave enough to try bad candy. And there isn't no feature for the model to detect this
